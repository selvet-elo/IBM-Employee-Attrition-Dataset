In this notebook, exploratory data analysis will be conducted to understand the data and the relationships between the variables.


```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from pathlib import Path

file_path = Path("C:/Users/selo/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df = pd.read_csv(file_path)
df.head(5)
```

### Exploratory Data Analysis:

```python
df.info()
```

```python
df.select_dtypes(include=['int64','float']).describe().T
```

```python
df.select_dtypes(include=['object']).describe().T
```

Value counts for categorical variables 

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f"Value counts for column '{col}':\n{df[col].value_counts()}\n")
    print('-----------------------')
```

For each column, unique value counts:  

```python
for column in df.columns:
    print(f'{column}:Unique value number: {df[column].nunique()}')
    print("-------------")
```

We can see some of these columns ('EmployeeCount', 'Over18', 'StandardHours') have only one unique value, those are not helpful for our analysis. We can drop these columns. Also, we can drop the 'EmployeeNumber' column beacuse it doesn't have any valuable input to our analysis. 

```python
df.drop(['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'], axis=1, inplace=True)
```

Let's define a function to create a count plot that can be used both individually and within a subplot.

```python
def dist_plot(df, x_col, hue_col, ax=None, order=None, ticks_rot=0):
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 5))
    count_plt = sns.countplot(data=df, x=x_col, hue=hue_col, palette='rocket', ax=ax, order=order)
    ax.set_xlabel(x_col)
    ax.set_ylabel('Count')
    ax.set_title(f'{hue_col} Distributions for {x_col}')
    for bars in count_plt.containers:
        ax.bar_label(bars, fmt='%d', label_type='edge', padding=2)
    ax.tick_params(axis='x', rotation=ticks_rot)   
```

## Quick observation on the plots above
When we examine these four plots, we can see that they primarily provide insights into employee satisfaction. Therefore, I decided to analyze them together. 

The first noticeable distinction in these plots is that employees who work overtime have a significantly higher attrition rate.

Looking at job and environment satisfaction, as expected, attrition rates increase as satisfaction levels decrease.

Regarding work-life balance, the most striking observation is that very few employees rate their work-life balance as "Best," while the majority fall under the "Better" category. Employees with a "Better" work-life balance have notably low attrition rates, whereas those who report a poor work-life balance experience much higher attritionâ€”almost twice as much.

Considering all these factors, we can generalize that lower work satisfaction significantly contributes to employees leaving their jobs.
